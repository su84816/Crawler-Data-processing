```python
import requests
import codecs

url = 'https://www.mol.gov.tw/'
url2 = 'https://www.youtube.com/'
url3 = 'https://www.104.com.tw/jobs/main/'
#requests主要用法,加入更多header資訊可以騙過server,減少被檔
r1 = requests.get(url,headers={
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:93.0) Gecko/20100101 Firefox/93.0'
}
                  )

#print(r1.status_code) #印狀態碼
# for k,v in r1.headers.items(): #把headers裡面的字典印出來
#  print(k,v)
#print(r1.encoding) #讀編碼
#r1.encoding='Big5'#如果讀不到編碼,測試網頁是不是用BIG5寫的,改預設編碼,通常中文用Big5/utf-8
# print(r1.encoding)
# print(r1.text) #讀html
with codecs.open('/content/drive/MyDrive/AI_02/1.html','w',r1.encoding) as f: #第一個變數一開始用1.html建檔,完成drive指向後可以改路徑"/content/drive/MyDrive/AI_09/1.html",AI_09前面是固定的
  f.write(r1.text) 
```
