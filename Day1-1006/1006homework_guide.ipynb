```python
import requests
import codecs
url = 'http://teaching.bo-yuan.net/test/requests/'
#requests主要用法,加入更多header資訊可以騙過server,減少被檔
#####改用法 #####3:get>delete #####5:改put #####7:改patch #####9:改post
r1 = requests.post(url,
  headers={
      'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:93.0) Gecko/20100101 Firefox/93.0',
      'action':'action' 
  },
  #####2:加入參數-action
  params={
     'action':'action',
  },
  data = {
      'id':'id', #####4:加資料-id
      'name':'name', #####6:加資料name
      'address':'address' #####8:加資料"address""
  },
          )

#print(r1.encoding) #編碼是ISO-8859-1
# print(r1.content)
r1.encoding = 'utf-8'  #####1:改編碼
print(r1.text)
```
