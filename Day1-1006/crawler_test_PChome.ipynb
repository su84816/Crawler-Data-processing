```python
#目標url
#######################ask：原網址為滾動式換頁，非手動換頁的狀況下，page是否有效?
url = 'https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=iphone13%20pro&page={page}&sort=sale/dc'
#原本網址去抓：https://ecshweb.pchome.com.tw/search/v3.3/?q=iphone13%20pro 用f12>ctrl+f>找關鍵字>找到關鍵字在網站架構下,找到url的tittle
#目標url:https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=iphone13%20pro&page=1&sort=sale/dc
#https://ecshweb.pchome.com.tw/search/v3.3/?q=ipad
```

```python
#看網站結構及存取的型態,取出關鍵值,
#可以用try except來抓錯
for i in range(1,10):
  print(f'-----------------------{i}----------------------------')
  print(url.format(page=i))
  res = requests.get(url.format(page=i))
  pchome = res.json() #內建可以讀json的排版
  for prod in pchome['prods']:
    print(prod['name'])

```
