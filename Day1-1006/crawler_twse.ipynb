```python
#written by 明恩
import requests, json  
import prettytable

r1=requests.get("https://www.twse.com.tw/rsrc/data/zh/home/yields.json",params={"_":"163366517423"})
d=json.loads(r1.text)

t = [0, 1, 2, 3]
 
for k in range(1,4,1):
  t[k] = prettytable.PrettyTable(["country","cost"], encoding="utf8") 

for i in range(1,4,1):
  print(d["chart1"]["table"+str(i)]["title"])
  for a in d["chart1"]["table"+str(i)]["data"]:
    t[i].add_row([a[0],a[1]])
  
  print(t[i])
```

```python
import requests, json
import prettytable

ur=requests.get("https://www.twse.com.tw/exchangeReport/STOCK_DAY",
    params={
        "response":"json",
        "date":"20210901",
        "stockNo":"2603",
     }     
)

d = json.loads(ur.text)

t1=prettytable.PrettyTable([d["fields"][0],d["fields"][1],
              d["fields"][2],d["fields"][3],
              d["fields"][4],d["fields"][5],
              d["fields"][6],d["fields"][7],
              d["fields"][8]], encoding="utf8")

print(d["title"])

for a in d["data"]:
    t1.add_row([a[0],a[1],a[2],a[3],a[4],a[5],a[6],a[7],a[8]])

t1.align[d["fields"][0]]="l"
t1.align[d["fields"][1]]="r"
t1.align[d["fields"][2]]="r"
t1.align[d["fields"][3]]="r"  
t1.align[d["fields"][4]]="r"
t1.align[d["fields"][5]]="r"
t1.align[d["fields"][6]]="r"
t1.align[d["fields"][7]]="r"
t1.align[d["fields"][8]]="r"

print(t1)
```
