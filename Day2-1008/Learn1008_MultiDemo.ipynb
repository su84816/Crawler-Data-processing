```python
import requests
import csv
import io

url = 'https://data.taipei/api/getDatasetInfo/downloadResource?id=e448bdbd-fd18-40fc-8f0e-a2ff4ce44cf4&rid=78068a2e-e72a-4ee8-b4b4-bcbd7a12d913' #opendata
r1 = requests.get(url) #對opendate做資料讀取,拿到網址內的資料
# print(r1.text) #如果跑亂碼,要先encoding改編碼
r2 = io.StringIO(r1.text) #把r1.text做字串解析
r3 = list(csv.reader(r2)) #再把r2做list處理+用csv去讀
#老師寫法(直接做完): d = list(csv.reader(io.StringIO(r1.text)))
#如果對象csv中有字串要用到",",通常會用雙引號把字串包起來
print(r3) #印出"\ufeff"代表使用編碼:utf-8,但要看輸出資料者有沒有先加註
#這個list是一個二維list,意思是list內包list,2D/3D的概念 ####json會有多維
####遇到SSL Error , https代表SSL加密過期 , 在url後面加上verify=False就可以正常讀取 , 但會多跑一行路徑內容
for i in r3:
  if len(i) > 3: #避免超出索引,第三行的內容是空值,可能第3欄裡面有空格 在IndexError: list index out of range處理
    print(i[0],i[3]) #把r3-list裡面的第0個和第3個印出(第0欄&第3欄)
  # print(r3[0],r3[3])
```

```python
import requests
import json

url = 'https://datacenter.taichung.gov.tw/swagger/OpenData/997d7dfb-3d93-437c-bb57-8e5dd027cb79' #opendata
r1 = requests.get(url) #對opendate做資料讀取

# print(r1.text) #如果跑亂碼,要先encoding改編碼
j1 = json.loads(r1.text)
#print(j1) #先看讀到的資料格式,或是用firefox查看
for a in j1: 
  print(a["金額1"],a["金額2"]) #格式是list包字典,所以取值要用字典去下指令
```

```python
#多層範例,要看懂dict
import requests,json
r1=requests.get("https://api.1922.gov.tw/vab-opendata/api/waitingAmountByCity")
d=json.loads(r1.text)
for a in d["result"]: #多層處理,用for去跑裡面的串列,所以會從index:0這層開始 #d後面接第一層
  print(a["city"],a["vaccine"]["高端"],a["vaccine"]["BNT"],a["vaccine"]["AZ"]) #抓city的值 #抓vaccine的值 裡面(又包一層字典)高端的值.... #a後面接第二層 #有第三層就直接接在a[]後面
```

```python
r1=requests.get("https://datacenter.taichung.gov.tw/swagger/OpenData/997d7dfb-3d93-437c-bb57-8e5dd027cb79")
d=json.loads(r1.text)
for i in range(1,8,1): #把標題印出來
  print("金額"+str(i),end="\t")
print("") #換行
for a in d:
  for i in range(1,8,1): #跑7次,把金額1-7用str(i)代換印出來
    print(a["金額"+str(i)],end="\t") #金額+i,i=1~7 \t=tab
  print("") #換行
```

```python
#以udn API串接為例,原網址:https://udn.com/news/breaknews/1 ， 
# 1.首先要先抓API-URL ， 開著F12往下轉會自己讀API的資料(看網頁，或是有的是會用F5) ， 反白選取關鍵字後 ， 在"全部"的狀況下用放大鏡搜尋"關鍵字"(用COPY的) ，會跳出關鍵字所屬的網域位置"udn.com" (你可以用過濾網指加入"udn.com")，
# 2.按檢測>XHR篩選 ， 在回應裡可以找到你要的新聞資料(要一個一個找，你可以在右邊過濾屬性的位置加入"關鍵字"，可以更快速找到) ， 就可以找到API的URL
#老師重新示範
# 1.載入,等他跑完
# 2.清空 下拉,跑API資料
# 3.往下拉 看回應裡面哪裡有新聞標題資料 #中文找不到的話可能是經過編碼了 找英文或數字(時間)去搜 #或是你先找UTF-8的編碼出來搜(要是16進位制) #用小算盤可以算進位制的轉換(左上角程式設計人員)>調位元
# 4.找到有API的URL 

for p in range(1,5,1): #page=1~4
  r1=requests.get(
    "https://udn.com/api/more", #原url：https://udn.com/api/more?page=2&id=&channelId=1&cate_id=0&type=breaknews&totalRecNo=13531 ，從?開始後面是變數，用&做區分並丟到params，這樣不會有編碼的問題
    params={
      "page":p, #這裡做頁碼自動跌加1~4
      "id":"",
      "channelId":"1",
      "cate_id":"0",
      "type":"breaknews",
      "totalRecNo":"13531" #就算是數字也要放字串,比較安全
    }
  )
  d=json.loads(r1.text)
  for a in d["lists"]:
    print(a["time"]["date"],a["title"])
```

```python
#以證交所為例：https://www.twse.com.tw/zh/ ，進去開著f12重新整理， 用放大鏡搜尋關鍵字"新加坡" ， 會找到他的檔案位置關鍵字"yields.json" ， 在XHR中找到檔案位置後 ，在檔頭就可以找到URL , 在回應中找到所顯示的資料
#URL:https://www.twse.com.tw/rsrc/data/zh/home/yields.json?_=1633667549052 ， 在?後方是變數(只有一個)：'_' : '163366517423'
r1=requests.get("https://www.twse.com.tw/rsrc/data/zh/home/yields.json",params={"_":"163366517423"})
d=json.loads(r1.text)
for i in range(1,4,1):
  print(d["chart1"]["table"+str(i)]["title"])
  for a in d["chart1"]["table"+str(i)]["data"]:
    print(a[0],a[1])
  print("")
```

```python
import prettytable
import requests
import json
r1=requests.get("https://datacenter.taichung.gov.tw/swagger/OpenData/997d7dfb-3d93-437c-bb57-8e5dd027cb79")
d=json.loads(r1.text)
title=[]
for i in range(1,8,1): #這裡在做title的建立,金額1-7
  title+=["金額"+str(i)]
  t1=prettytable.PrettyTable(title, encoding="utf8")
for a in d:
  data=[]
  for i in range(1,8,1):
    data+=[a["金額"+str(i)]]
  t1.add_row(data)
print(t1)
```
