```python
import requests
import prettytable
import json
#以iphone為例，原網址：https://ecshweb.pchome.com.tw/search/v3.3/?q=iphone
#找到後的url：https://ecshweb.pchome.com.tw/search/v3.3/all/results?q=iphone&page=1&sort=sale/dc
url = 'https://ecshweb.pchome.com.tw/search/v3.3/all/results'



input_page = 1 #預設頁碼
input_search = input("請輸入你要搜尋的關鍵字:")#####這裡要補防呆
input_sort = "sale/dc"
input_sort = input("你想依據什麼排序方式?([預設]有貨優先:A),(精準度:B),(價錢由高至低:C),(價錢由低至高:D)(新上市:E)")######這裡要補防呆


A = "sale/dc" #有貨
B = "rnk/dc"  #精準
C = "prc/dc"  #價錢,遞減
D = "prc/ac"  #價錢,遞增
E = "new/dc"  #新上市



while True:

  res = requests.get(url,
        params={"q":input_search,"page":input_page,"sort":input_sort}, 
        headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:93.0) Gecko/20100101 Firefox/93.0"},
                         )
  #####加入無法連線時出現通知,from益詠
  if str(res.status_code)[0] !="2": #如果狀態碼開頭不是2就跳出(代表網頁狀態有問題)    
    print("網頁狀態有問題,請檢查程式碼!")    
    break
  j = json.loads(res.text)
  #####Beta:找不到頁碼的狀況>比對找得到和找不到的內容 + 荷雅提供totalPage
  total_page = int(j['totalPage'])
  print(f'本次搜尋,總頁數：{total_page}')
  print(f'這是第{input_page}/{total_page}頁')

  t1 = prettytable.PrettyTable(["名稱","價格"], encoding="utf8") 
  t1.align["名稱"]="l"
  t1.align["價格"]="c" 
   
  for a in j["prods"]: #多層處理,用for去跑裡面的串列,所以會從index:0這層開始 #d後面接第一層
    t1.add_row([a["name"],a["price"]]) #用一個list去包name&price的字串
  print(t1)
  t1.clear() #這裡在做TableList的清空,讓迴圈的資料不會一直疊加

  input_page = int(input("請輸入頁碼:")) #####這裡要補防呆
  if input_page > total_page:
    input_page = int(input(f'頁碼超出範圍,請重新輸入{total_page}以下數字:'))
  #####我想加一個防呆計時器,使用者輸入時間超過10秒就停止程式

# for a in j["prods"]: #多層處理,用for去跑裡面的串列,所以會從index:0這層開始 #d後面接第一層
#   print(a["name"],a["price"])
  



```
