```python
import requests
from bs4 import BeautifulSoup
import prettytable

url = 'https://www.cwb.gov.tw/V8/C/W/TemperatureTop/County_TMax_T.html?'
r1 = requests.get(url,
        # params = {
        #     "ID":"=Wed Oct 10 2021 13:25:42 GMT 0800 (台北標準時間)"
        # },
        headers = {
            "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:93.0) Gecko/20100101 Firefox/93.0",
              }
          )
b1 = BeautifulSoup(r1.text,"html.parser") #對r1.text拿到的內容作html分析
# print(b1) #先測試能不能連線

t1 = prettytable.PrettyTable(["地區","氣溫"], encoding="utf8") 
t1.align["地區"]="l"
t1.align["氣溫"]="c" 

for b2 in b1.find_all("tr"): 
  t1.add_row([b2.find("th",{"scope":"row"}).text,b2.find("span",{"class":"tem-C is-active"}).text])
print(t1)
  # print(
  #     b2.find("th",{"scope":"row"}).text, #這裡取縣市
  #     b2.find("span",{"class":"tem-C is-active"}).text #這裡取溫度
  #     ) 
                
```
