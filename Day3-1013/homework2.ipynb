```python
import requests
import io
import json
import prettytable #import整個函示庫


input_search=(input('請輸入查詢物品:'))
input_page = 1
input_sort = "sale/dc"
# t1=PrettyTable() #先註解掉,不要預設值

while True:

  ur=requests.get("https://ecshweb.pchome.com.tw/search/v3.3/all/results",
                params={
                "q":input_search,
                "page":input_page,
                "sort":input_sort          
            },
              headers = {
                  "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:93.0) Gecko/20100101 Firefox/93.0"
              }          
  )

  data  = json.loads(ur.text)
  total_page=int(data['totalPage'])
  print(f'本次搜尋,總頁數{total_page}')

  t1=prettytable.PrettyTable(["名稱","價格"], encoding="utf8")
  t1.align['名稱']='l'
  t1.align['價格']='c'

  for a in data['prods'] :
    t1.add_row([a['name'],a['price']]) #變數修正為t1,要改可以改,但要注意所有的變數都要一起改,這裡a的中括號'['要在a的前面,price前面也要加a
    ###意思是:t1.add_row 我要加值進去這一列,要加什麼值?(你要給他list)>>> [a['name'],a['price']] <這是一個list,裡面包a['name'] & a['price']
  print(t1)
  t1.clear() #變數修正為t1,要改可以改,但要注意所有的變數都要一起改

  input_page  = int(input('請輸入頁碼:'))
  if input_page > total_page:
    input_page  = int(input(f'頁碼超出範圍，請重新輸入{total_page}以下數字:'))
```
